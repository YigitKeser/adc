```python
import sys
import os
from image_processing import generate_image
from ad_template import create_ad_template

def main(image_path, text_prompt, hex_color, punchline_and_button_color_hex, logo_path, punchline_text, button_text):
    
    result_image = generate_image(image_path, text_prompt, hex_color)
    
    if not os.path.exists('output'):
        os.makedirs('output')
    result_image.save("output/generated_image.png") 

    base_image_path = "output/generated_image.png"
    ad_image = create_ad_template(base_image_path, logo_path, punchline_and_button_color_hex, punchline_text, button_text)
    ad_image.save("output/ad_template.png")
    return result_image, ad_image
```

```python
text_prompt = "take away cup with coffee,medium size table,christmas tree at background,snow from the window"
hex_color = "#50C878"
punchline_and_button_color_hex = "#50C878"
image_path = "input/test_image.png"
logo_path = "input/logo.png"
punchline_text = "Enjoy the best coffee in town!"
button_text = "Get your Free Coupon"

result_image, ad_image = main(image_path, text_prompt, hex_color, punchline_and_button_color_hex, logo_path, punchline_text, button_text)

result_image
```
