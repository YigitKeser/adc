```python
import requests
import time
from PIL import Image

def submit_ad_request_and_download(image_file_path, logo_file_path, text_prompt, hex_color, punchline_and_button_color_hex, punchline_text, button_text):
    # URL of the FastAPI endpoint
    url = "http://127.0.0.1:8000/generate-ad/"
    status_url = "http://127.0.0.1:8000/tasks/"
    download_url = "http://127.0.0.1:8000/download/"

    # Prepare the data for the form fields
    data = {
        'text_prompt': text_prompt,
        'hex_color': hex_color,
        'punchline_and_button_color_hex': punchline_and_button_color_hex,
        'punchline_text': punchline_text,
        'button_text': button_text
    }

    # Prepare the files for upload
    files = {
        'image': (image_file_path, open(image_file_path, 'rb')),
        'logo': (logo_file_path, open(logo_file_path, 'rb'))
    }

    # Send the POST request to generate the ad
    response = requests.post(url, data=data, files=files)

    # Close the files after sending the request
    files['image'][1].close()
    files['logo'][1].close()

    # Get the task ID from the response
    task_id = response.json().get("task_id")

    # Check the status of the task
    while True:
        status_response = requests.get(f"{status_url}{task_id}")
        status = status_response.json().get("status")

        if status == "completed":
            print("Task completed. Downloading the image...")
            download_response = requests.get(f"{download_url}{task_id}")

            with open('generated_ad.png', 'wb') as f:
                f.write(download_response.content)
            print("Image downloaded as 'generated_ad.png'")
            break

        elif status == "invalid task id":
            print("Invalid task ID.")
            break

        print("Task is still processing...")
        time.sleep(5)  # Wait for a few seconds before checking again
```

```python
submit_ad_request_and_download(
    "test_image.png",
    "logo.png",
    "take away cup of a coffee with cover on a table with some flowers background",
    "#50C878",
    "#50C878",
    "Enjoy the best coffee in town!",
    "Get your Free Coupon"
)

image = Image.open('generated_ad.png')
image
```
